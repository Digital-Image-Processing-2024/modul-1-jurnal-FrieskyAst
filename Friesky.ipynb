```python
image = plt.imread("aang.jpg")
ang = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
plt.imshow(ang,cmap='gray')
```

```python
def hist(ang, N, M):
    hist =[0]*255
    for i in range (255):
        hist[i]=0
    for i in range(N):
        for j in range(M):
            hist[ang[i][j]]+=1
    n= N*M
    for i in range(255):
        hist[i] = hist[i]/float(n)
    return hist
        
```

```python
def rotate(image, angle):
    (h, w) = image.shape[:2]
    center = (w // 2, h//2)
    M = cv.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv.warpAffine(image, M, (w, h))
    return rotated
```

```python
image = cv.imread('aang.jpg')
angle = 40
rotated =rotate(image, angle)
plt.imshow(rotated)
```
