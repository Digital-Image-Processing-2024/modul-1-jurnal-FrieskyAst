## HISTOGRAM

Kode menggunakan perulangan 'for' untuk mengisi list 'hist' dengan 255 elemen dengan nilai 0. Lalu, menggunakan perulangan 'for' untuk melakukan iterasi setiap pixel gambar angka. Dalam perulangan, menggunakan perulangan lain untuk melakukan iterasi setiap elemen matriks 'ang'. Kemudian, menggunakan perulangan 'if' untuk memastikan bahwa nilai pixel tidak melewati batas 255. Jika nilai pixel valid, maka menambahkan 1 pada elemen list 'hist' yang sesuai dengan nilai pixel. Setelah itu, menggunakan perulangan 'for' untuk menghitung persentase setiap elemen list 'hist'. Lalu, menggunakan perulangan 'for' untuk menghitung jumlah pixel gambar angka dengan 'N*M'. Kemudian, menggunakan perulangan 'for' untuk menghitung persentase setiap elemen list 'hist' dengan rumus 'hist[i] = hist[i]/float(n)'. Setelah iyu, kode mengembalikan list 'hist' sebagai hasilnya.

```python
def hist(ang, N, M):
    hist =[0]*255
    for i in range (255):
        hist[i]=0
    for i in range(N):
        for j in range(M):
            hist[ang[i][j]]+=1
    n= N*M
    for i in range(255):
        hist[i] = hist[i]/float(n)
    return hist
```

## MENYIMPAN KERNEL

Kode berikut digunakan untuk menyimpan kernel yang kemudian akan di pakai pada konvolusi

```python
k = np.array([[1, 2, -1, 1],
                   [0, 2, 1, 2],
                   [2, 1, 1, 0],
                   [1, 0, -1, 0]])

h, w = k.shape
```

```python
h, w = ang.shape
kernelH, kernelW = k.shape

padsize = int((kernelH - 1) / 2)
padW = ((padsize, padsize+1), (padsize, padsize+1))
padding_grayimage = np.pad(ang, padW, mode="constant")
zeropad = np.zeros((h, w))

for i in range(zeropad.shape[0]):
    for j in range(zeropad.shape[1]):   
        subset = padding_grayimage[i:i + kernelH, j:j + kernelW]
        LebarSub = subset * k
        sumSub = np.sum(LebarSub)
        zeropad[i][j] = np.round(sumSub)
print(zeropad)
```

# TRANSLASI DAN ROTASI

## ROTASI

Sebelum menampilkan hasil rotasi terdapat perhitungan yang harus dilakukan. Dalam hal tersebut pada paragraf pertama, mengambil ukuran gambar angka dengan variabel 'h' yaitu tinggi dan 'w' yaitu lebar. Lalu, membuat sebuah matriks nol berukuran 'h x w' dengan tipe data 'int' dengan variabel 'roted'. Setelah itu, mengubah sudut 40 derajat ke sudut radian dengan 'np.deg2rad(40)'. Kemudian, mengambil nilai cosinus dan sinus dari sudut radian dengan 'np.cos(angle)' dan 'np.sin(angle)' yang dimana 'angle' merupakan sudut rotasi.
Kemudian pada paragraf kedua, mengambil tinggi dan lebar gambar angka dengan variabel 'tinggi' dan 'lebar'. Lalu, menggunakan perulangan 'for' untuk melakukan iterasi setiap elemen matriks 'roted'. Dalam perulangan, menggunakan rumus rotasi untuk menghitung koordinat rotasi 'rotasiX' dan 'rotasiY'. Kemudian, mengubah koordinat rotasi ke koordinat geser 'slideX' dan 'slideY' dengan menambahkan 'tinggi' dan 'lebar' pada koordinat rotasi. Kemudian, menggunakan perulangan 'if' untuk memastikan bahwa koordinat slide tidak melewati batas gambar angka. Jika koordinat slide valid, kode mengisi matriks 'roted' dengan nilai dari matriks 'ang' yang merupakan gambar grayscale.

```python
h, w = ang.shape
roted = np.zeros((h, w)).astype(int)
angle = np.deg2rad(40)
cos = np.cos(angle)
sin = np.sin(angle)

tinggi = int((h - 1) / 2)
lebar = int((w - 1) / 2)

for i in range(roted.shape[0]):
    for j in range(roted.shape[1]):
        rotasiX = (i - tinggi) * cos - (j - lebar) * sin
        rotasiY = (i - tinggi) * sin + (j - lebar) * cos
        
        slideX = int(rotasiX + tinggi)
        slideY  = int(rotasiY + lebar)
        
        if 0 <= slideX < h and 0 <= slideY < w:
            roted[i][j] = ang[slideX][slideY]
```

## MENAMPILKAN ROTASI

```python
plt.imshow(roted, cmap='gray')
```

Dari perhitungan yang telah dilakukan untuk merotasikan gambar, maka dengan menggunakan 'imshow' dengan memanggil 'roted' yang merupakan perhitungan rotasi dan 'cmap='gray'' yang merupakan hasil dari merubah warna rgb menjadi grayscale makan akan tertampil hasil rotasi 40 derajat

## TRANSLASI

Kode menggunakan fungsi np.zeros() untuk membuat matriks nol dengan ukuran (h, w) dan tipe data int. Lalu, kode menggunakan perulangan for untuk melakukan iterasi setiap pixel gambar roted. Dalam perulangan, menggunakan perulangan lain untuk melakukan iterasi setiap elemen matriks roted. Kemudian, menggunakan variabel tX dan tY untuk menghitung koordinat yang akan digunakan untuk mengisi matriks t. Kemudian, menggunakan perulangan if untuk memastikan bahwa koordinat tX dan tY tidak melewati batas gambar roted. Jika koordinat valid, kode mengisi matriks t dengan nilai dari matriks roted yang sesuai dengan koordinat tX dan tY. Setelah itu, kode menggunakan fungsi plt.imshow() dengan argumen t dan cmap='gray' untuk menampilkan gambar matriks t.

```python
t = np.zeros((h, w)).astype(int)

for i in range(h):
    for j in range(w):
        tX = i - 300
        tY = j + 163
        
        if 0 <= tX < h and 0 <= tY < w:
            t[i][j] = roted[tX][tY]
```

## MENAMPILKAN TRANSLASI

```python
plt.imshow(t, cmap='gray')
```

Dari perhitungan yang telah dilakukan untuk mentranslasi gambar, maka dengan menggunakan 'imshow' dengan memanggil 't' yang merupakan perhitungan translasi dan 'cmap='gray'' yang merupakan hasil dari merubah warna rgb menjadi grayscale makan akan tertampil hasil translasi.

# PENCERMINAN DAN DILATASI

## PENCERMIANAN TERHADAP SUMBU Y

Kode menggunakan fungsi np.zeros() untuk membuat matriks nol dengan ukuran (h, w) dan tipe data int. Lalu, menggunakan perulangan for untuk melakukan iterasi setiap pixel gambar t. Dalam perulangan, kode menggunakan perulangan lain untuk melakukan iterasi setiap elemen matriks t. Kemudian, menggunakan variabel w - j - 1 untuk menghitung koordinat yang akan digunakan untuk mengisi matriks refleksi. Kemudian, menggunakan perulangan if untuk memastikan bahwa koordinat tidak melewati batas gambar t. Jika koordinat valid, kode mengisi matriks refleksi dengan nilai dari matriks t yang sesuai dengan koordinat.

```python
h, w = ang.shape
refleksi = np.zeros((h, w), dtype=int)
for i in range(h):
    for j in range(w):
        refleksi[i][j] = t[i][w - j - 1]
```

## MENAMPILKAN PENCERMINAN

```python
plt.imshow(refleksi, cmap="gray")
```

Dari perhitungan yang telah dilakukan untuk merotasikan gambar, maka dengan menggunakan 'imshow' dengan memanggil 'refleksi' yang merupakan perhitungan pencerminan dan 'cmap='gray'' yang merupakan hasil dari merubah warna rgb menjadi grayscale makan akan tertampil hasil refleksi terhadap sumbu Y.

## DILATASI

```python
d = np.zeros((h*3, w*3), dtype=int)
for i in range (h*3) :
    for j in range (w*3) :
        newH = int(h * i / (h*3))
        newW = int(w * j / (w*3))
        d[i][j] =  refleksi[newH][newW]
```

## MENAMPILKAN HASIL DILATASI

```python
plt.imshow(d,cmap="gray")
```

Dari perhitungan yang telah dilakukan untuk merotasikan gambar, maka dengan menggunakan 'imshow' dengan memanggil 'd' yang merupakan perhitungan 
dilatasi dan 'cmap='gray'' yang merupakan hasil dari merubah warna rgb menjadi grayscale makan akan tertampil hasil dilatasi yang terbuhung dengan refleksi terhadap sumbu Y.
